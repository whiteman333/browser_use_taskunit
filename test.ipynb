```python
from langchain_openai import ChatOpenAI
from PIL import Image
import base64
from io import BytesIO

llm = ChatOpenAI(
    model = "Qwen/Qwen2.5-72B-Instruct-GPTQ-Int8",
    temperature = 0,
    base_url = "http://localhost:12111/v1",
    api_key = "empty",
    max_tokens = 512
)

image = Image.open("/home/zyy/browser_use_naive/last_screenshot.png")

# Convert image to base64 string
buffered = BytesIO()
image.save(buffered, format="PNG")
img_str = base64.b64encode(buffered.getvalue()).decode()

# Create message with image
messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "text", 
                "text": "Please describe what you see in this image:"
            },
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/png;base64,{img_str}"
                }
            }
        ]
    }
]

# Get response from LLM
response = llm.invoke(messages)
print(response.content)
```

```python
from langchain_openai import ChatOpenAI

task = "在西南科技大学计算机科学与技术学院范勇院长为2024级新生进行的入学教育讲座中，他引用了哪几位名人以激励学生？"

PROMPT = """
You are a helpful Query Rewriter.

You are given a task and a list of important pages.

Please output several given important pages in the list that can mostly complete the task.
"""

llm = ChatOpenAI(
    model = "/home/zyy/LLMs/Qwen/Qwen2.5-VL-72B-Instruct-AWQ",
    temperature = 0,
    base_url = "http://localhost:12111/v1",
    api_key = "empty",
    max_tokens = 512
)

# Create message with image
messages = [
    {
        "role": "system",
        "content": PROMPT
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text", 
                "text": "Important Links: " + webwiki.retrieve_important_links(threshold = 2)
            }
        ]
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text", 
                "text": "Task: " + task
            }
        ]
    }
]

response = llm.invoke(messages)
print(response.content)
```
