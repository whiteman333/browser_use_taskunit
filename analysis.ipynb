```python
# Load the task file
import json
task_path = '/home/zyy/web_analyzer/output/low_count_filtered_tasks.json'
with open(task_path, 'r') as f:
    tasks = json.load(f)

# Count tasks per website
website_counts = {}
for task in tasks:
    website = task['website']
    website_counts[website] = website_counts.get(website, 0) + 1

# Sort websites by task count in descending order
sorted_websites = sorted(website_counts.items(), key=lambda x: x[1], reverse=True)

# Select top websites until we have 100 tasks
selected_websites = []
total_tasks = 0
for website, count in sorted_websites:
    selected_websites.append(website)
    total_tasks += count
    if total_tasks >= 100:
        break

print("\nWebsite task counts (top websites containing first 100 tasks):")

for website, count in sorted_websites[:len(selected_websites)]:
    print(f"{website}: {count} tasks")

print(f"\nTotal websites selected: {len(selected_websites)}")
print(f"Total tasks in selected websites: {total_tasks}")

# Save selected tasks to a new file
selected_tasks = []
for task in tasks:
    if task['website'] in selected_websites:
        selected_tasks.append(task)

# Save to new JSON file
output_path = '/home/zyy/web_analyzer/output/low_count_filtered_tasks_100.json'

# Sort tasks by website frequency in descending order
website_task_counts = {}
for task in selected_tasks:
    website = task['website']
    website_task_counts[website] = website_task_counts.get(website, 0) + 1

# Sort tasks based on their website's frequency count
selected_tasks.sort(key=lambda x: website_task_counts[x['website']], reverse=True)

# Save sorted tasks back to file
with open(output_path, 'w') as f:
    json.dump(selected_tasks, f, indent = 2)

print(f"\nSaved {len(selected_tasks)} tasks sorted by website frequency to {output_path}")
```
